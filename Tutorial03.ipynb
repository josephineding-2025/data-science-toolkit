```python
import numpy as np
```

```python
new=np.array([[0,1],[2,3]])
print(new)
```

```python
random_array=np.random.rand(3,4)
print(random_array)
```

```python
new2=np.ones([2,2])
new3=np.zeros([2,2])
print(new2)
print(new3)
```

```python
q4=np.random.rand(6,3)
new4=np.transpose(q4)
print(q4)
print(new4)
```

```python
n=2
p=3
m=4
random1=np.random.rand(n,p)
random2=np.random.rand(p,m)
print(random1,"\n",random2)

dot_product=np.dot(random1,random2)
print(dot_product)
```
